```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt
from matplotlib import dates
```

```python
df = pd.read_csv('datasets/Chapter 1/trip.csv')
df.head()
```

```python
#Ex 01
List_ = list(df['starttime'])
List_ = [datetime.datetime.strptime(x, "%m/%d/%Y %H:%M") for x in List_]
df['starttime_year'] = pd.Series([x.year for x in List_], index=df.index)

df_gbyear = df.groupby('starttime_year').count()
plt.bar(df_gbyear.index, df_gbyear['trip_id'])
plt.show()
```

```python
#EX 0 2
df['starttime_month'] = pd.Series([x.month for x in List_], index=df.index)
df_gpmonth = df.groupby('starttime_month').count()

plt.plot(df_gpmonth.index, df_gpmonth['trip_id'])
plt.show()
```

```python
#EX 03
df['starttime_day'] = pd.Series([x.day for x in List_], index=df.index)
df_gbday = df.groupby('starttime_day').count()

plt.plot(df_gbday.index, df_gbday['trip_id'])
plt.show()
```

```python
#EX 04
df['starttime_date'] = pd.Series([x.date() for x in List_], index=df.index)
df_gbdate = df.groupby('starttime_date').count()

plt.plot(df_gbdate.index, df_gbdate['trip_id'])
plt.show()
```

```python
#EX 05
df['week_day'] = pd.Series([x.weekday() for x in List_], index=df.index)
df_gbwd = df.groupby('week_day').count()
plt.figure(figsize=[10,5])
plt.bar(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], df_gbwd['trip_id'])
plt.show()
```
